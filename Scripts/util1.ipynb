```python
gwDfFil = gwDf[['STATION_NM','dec_lat_va','dec_long_v','DTdeparsed','WLA_DEM']].copy()
```

```python
gwDfFil.loc[gwDfFil['STATION_NM']=='JL-49-13-509']
gwDfFil['Date'] = pd.to_datetime(gwDfFil.DTdeparsed)
gwDfFil.drop(['DTdeparsed'], axis=1)
```

```python
stationDf = pd.DataFrame(columns=gwDfFil.keys())
```

```python
i=0
for station in pd.unique(gwDfFil.STATION_NM):
    #print(gwDfFil.loc[gwDfFil['STATION_NM']==station].iloc[0])
    stationDf = stationDf.append(gwDfFil.loc[gwDfFil['STATION_NM']==station].iloc[0], ignore_index=True)
    i+=1
```

```python
stationDf.dropna(inplace=True)
stationDf.describe()
```
